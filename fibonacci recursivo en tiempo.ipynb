```python
import time #modulo de tiempo
import numpy as np #modulo de arreglos y matrices
from matplotlib import pyplot as plt  #librearia de graficos
```

```python
#funcion que calcula   el fibonacci
#tomando como parametro al numero que se le aplica fib
def fib(x):
    a=0
    if(x == 0):
        return 0
    elif(x== 1):
        return 1
    else:
        return fib(x - 1) + fib(x - 2)
```

```python
#funcion calcula los valores de la funcion de fibonacci para todos los valores de n
#calcula el tiempo que toma cada valor de n para usar la funcion de fibonacci
def graf(n,tin):
    t=tin
    # n valores de iteracion 0 hasta n
    vx=[]
    #tiempos que tarda la funcion fib(n) para cada valor de n 
    vy=[]
    #valores de f(n) para valores de 0 hasta n
    vz=[]
    for i in range(n):
        a=fib(i)
        b=time.clock()-t
        vx.append(i)
        vy.append(b)
        vz.append(a)
        t=time.clock()
    return vx,vy,vz
```

```python
n=40
tiempo=time.clock()
print tiempo
x,y,z=graf(n,tiempo)
```

```python

plt.plot(x,y,'ko')
plt.legend(('tiempo de ejecucion fibonacci' ), loc = 'upper left')
plt.xlabel("numeros de fibonacci", fontsize = 20)
plt.ylabel("tiempo de ejecucion segundos", fontsize = 20)
plt.title("tiempo de ejecucion fibonacci", fontsize = 20)
```

```python
plt.plot(x,z,'ro')
plt.legend(('tiempo de ejecucion fibonacci' ), loc = 'upper left')
plt.xlabel(("fib de n"), fontsize = 20)
plt.ylabel("tiempo de ejecucion segundos", fontsize = 20)
plt.title("valores de fib de n vs n", fontsize = 20)
```
