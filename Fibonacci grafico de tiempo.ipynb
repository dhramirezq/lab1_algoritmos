```python
import time #modulo de tiempo
import numpy as np #modulo de arreglos y matrices
from matplotlib import pyplot as plt  #librearia de graficos
```

```python
#funcion que calcula   el fibonacci y el tiempo en el que tarda en ejecutar la funcion de fibonacci
#tomando como parametro al numero que se le aplica fib y el tiempo inicial tomado con time.time()
def fib(x,tin):
    a=0
    b=1
    for i in range(x):
        a,b=b, a+b
    tfin=time.clock()-tin
    return a,tfin

def graf(n,tin):
    t=tin
    vx=[]
    vy=[]
    vz=[]
    for i in range(n):
        a,b=fib(i,time.clock()-t+tin)
        vx.append(i)
        vy.append(b)
        vz.append(a)
        t=time.clock()
    return vx,vy,vz

    
```

```python
n=10000


print "el valor de n es %s al que se le calcula fibonacci" %(n)
tiempo=time.clock()
print time.clock()
x,y,z=graf(n,tiempo)
#print z


```

```python
plt.plot(y,'k')
plt.legend( ('tiempo de ejecucion fibonacci'), loc = 'upper left')
plt.xlabel("numeros de fibonacci", fontsize = 20)
plt.ylabel("tiempo de ejecucion segundos", fontsize = 20)
plt.title("tiempo de ejecucion fibonacci", fontsize = 20)
```
